```python
# -*- coding:utf-8 -*-
#basic feature engineering with the titanic data
import pandas as pd
import numpy as np
import string
```

```python
data_train = pd.read_csv('/home/cui/kaggle/train.csv')
print(data_train.columns)
```

```python
def subStringsInString(bigString, subString):
    for substring in subString:
#         print(type(substring), type(subString))
        if substring in bigString:
            return substring
        
    print(bigString)
    return np.nan
    
def replace_titles(data):
    title = data['Title']
    if title in ['Don', 'Major', 'Jonkheer', 'Rev', 'Col']:
        return 'Mr'
    elif title in ['Countess', 'Mme']:
        return 'Mrs'
    elif title in ['Mlle', 'Ms']:
        return 'Miss'
    elif title == 'Dr':
        if data['Sex'] == 'mail':
            return 'Mr'
        else: 
            return 'Mrs'
    else:
        return title
```

```python
title_list=['Mrs', 'Mr', 'Master', 'Miss', 'Major', 'Rev',
                    'Dr', 'Ms', 'Mlle','Col', 'Capt', 'Mme', 'Countess',
                    'Don', 'Jonkheer']
```

```python
#Make every one has a title
data_train['Title']=data_train['Name'].map(lambda x: subStringsInString(x, title_list))
# data_train['Title'] = map(lambda x: subStringsInString(x, title_list), data_train['Name'])
data_train['Title'] = data_train.apply(replace_titles, axis=1)
```

```python
print(data_train['Title'])
```

```python
# print(data_train['Cabin'])
for i in data_train['Cabin']:
    print(i)
```

```python
#Make the deck feature
cabin_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'Unknown']
data_train['deck'] = data_train['Cabin'].map(lambda x:subStringsInString(str(x), cabin_list))
```

```python
print(data_train['deck'])
```

```python
#make the feature Family size
data_train['Family_size'] = data_train['SibSp'] + data_train['Parch']
print(data_train['Family_size'])
```

```python
#make the feature Age*class
data_train['Age*Class'] = data_train['Age'] * data_train['Pclass']
print(data_train['Age*Class'])
```

```python
#make the feature Fare per person
data_train['Fare_per_person'] = data_train['Fare'] / (data_train['Family_size'] + 1)
print(data_train['Fare_per_person'])
```
